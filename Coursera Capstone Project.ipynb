# Clean the data into a single usable dataframe:

**Crime data cleaning**

```python
#take just the last 12 months of data available
crimedf = df_data_1[["Borough", "Major Category", "201709", "201710", "201711", "201712", "201801", "201802", "201803", "201804", "201805", "201806", "201807", "201808"]]
                    
crimedf.head()
```

```python
#set 'borough' as the index
crimedf = crimedf.set_index("Borough")
#drop area inconsistant with other data
crimedf = crimedf.drop('London Heathrow and London City Airports', axis=0)
#drop non-violent crimes
violentcrimes = ["Sexual Offences", "Violence Against the Person"]
crimedf = crimedf[crimedf["Major Category"].isin(violentcrimes)]
#create total crime for the year from the last 12 months
crimedf = crimedf.sum(axis = 1)

crimedf = crimedf.groupby('Borough').agg('sum')

crimedf.name = 'Total Crime'

crimedf.head()
```

# Scrape geo-coordinates of boroughs from wikipedia page

```python
#import packages
import requests
from bs4 import BeautifulSoup
#define URL for data scraping and set up BeautifulSoup
URL = "https://en.wikipedia.org/wiki/List_of_London_boroughs"
res = requests.get(URL).text
soup = BeautifulSoup(res,'lxml')

df_list = []
# for loop runs through the data scraped by BeautifulSoup and extracts long and lat data
for items in soup.find('table', class_= 'wikitable sortable').find_all('tr')[1::]:
    data = items.find_all(['td'])
    data1 = data[0]
    data8 = data[8]    
    try:
        borough_name = data1.get_text()
        borough_name = borough_name.split('[')
        borough_name = borough_name[0]
        borough_name = borough_name.strip()
        
        ll = data8.get_text()
        ll = ll.split('/')
        lat_long = ll[2]
        lat_long = lat_long.split('(')
        lat_long = lat_long[0]
        lat_long = lat_long.split(';')
        latitude = lat_long[0]
        latitude = latitude.strip()
        longitude = lat_long[1]
        longitude = longitude.strip()
        longitude = longitude.replace(u'\ufeff', '')
        latitude = float(latitude)
        longitude = float(longitude)

#       Append the borough name, latitude and logitude in a list
        df_list.append((borough_name, latitude, longitude))
    except IndexError:pass
```

```python
#format new dataframe with Longitude and Latitudes
df_boroughs = pd.DataFrame(df_list, columns=['Borough', 'Latitude' , 'Longitude'])

df_boroughs.head()
```

```python
#import all of the packages required to perform a linear regression in which the values are normalised and a machine learning train and test set can be used

from sklearn import linear_model
from sklearn import preprocessing
from sklearn.preprocessing import StandardScaler
import numpy as np
import pylab as pl

#define the variables which will be used
cdf = dfmerge[["Commutibilty Score", "Rate of good schools", "Pay (£)", "Average House Price (£)"]]
ind = ["Commutibilty Score", "Rate of good schools", "Pay (£)"]
dep = ["Average House Price (£)"]

#initilize the train and test sets using a numpy random function
msk = np.random.rand(len(cdf)) < 0.8
train = cdf[msk]
test = cdf[~msk]

#ensure the train set is converted to an array
x = np.asanyarray(train[ind])
y = np.asanyarray(train[dep])

#initialise the normalising function and execute it onto the train set
scaler = StandardScaler()
xnorm = scaler.fit_transform(x)
ynorm = scaler.fit_transform(y)

#define the linear regression function and fit our normilised train sets
regr = linear_model.LinearRegression()

regr.fit (xnorm, ynorm)
# The coefficients
print ('Coefficients: ', regr.coef_)
```

```python
#run the predict function on the test set
y_hat= regr.predict(test[ind])
#define the arrays of the test sets
x2 = np.asanyarray(test[ind])
y2 = np.asanyarray(test[dep])
#normalise the prediction and test sets
yhat2 = scaler.fit_transform(y_hat)
xnorm2 = scaler.fit_transform(x2)
ynorm2 = scaler.fit_transform(y2)

#output the residual sum of squares
print("Residual sum of squares: %.2f"
      % np.mean((yhat2 - ynorm2) ** 2))

#output explained variance score: 1 is perfect prediction
print('Variance score: %.2f' % regr.score(xnorm2, ynorm2))
```

```python
#define a plot which allows us to visualise how well the fitted values predit the actual values
width = 12
height = 10
plt.figure(figsize=(width, height))


ax1 = sns.distplot(ynorm2, hist=False, color="r", label="Actual Value")
sns.distplot(yhat2, hist=False, color="b", label="Fitted Values" , ax=ax1)


plt.title('Actual vs Fitted Values for Houses')
plt.xlabel('Price (£)')
plt.ylabel('')


plt.show()
```
